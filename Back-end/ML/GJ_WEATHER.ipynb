```python
import pandas as pd

weather = pd.read_csv('./data/weather.csv', encoding = 'euc-kr')
#weather.columns #Index(['지점', '지점명', '일시', '기온(°C)', '풍속(m/s)', '습도(%)'], dtype='object')
weather
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,8))
plt.plot(range(144), weather['기온(°C)'], marker='*', c='red', label='temperatuer')
plt.plot(range(144), weather['풍속(m/s)'], marker='*', c='green', label='wind')
plt.plot(range(144), weather['습도(%)'], marker='*', c='blue', label='humidity')
plt.title('GJ Weather(8/7~12)')
plt.legend(loc='upper right')
plt.grid()
plt.show()
```
